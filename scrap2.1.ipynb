```python
page_link=soup1.findAll('a',{'class':'ge-49M'})
domain = 'https://flipkart.com/q/best-laptops-under-rs-50000?page='
for i in range(2,36):
    link = domain+str(i)
    pagedata = urlopen(link)
    webdata1 = pagedata.read()
    webdata += webdata1
  
```

```python
soupdata = soup(webdata,'html.parser')
type(soupdata)
```

```python
containers = soupdata.findAll('div',{'class':'_2kHMtA'})
print(type(containers),len(containers))
```

```python
containers[0]  #data of first container
```

```python
import re

re.findall('\d+,*\d*','1,475 Ratings & 115 Reviews')
```

```python
import re

ratreview= re.findall('\d+,*\d*','1,475 Ratings & 115 Reviews')
print(ratreview[0],ratreview[1])
```

```python
f = open('laptopnew.csv','wb')
f.write('ProductName,Star,Ratings,Reviews,current_price,MRP,Processor,RAM,Storage,ImageURL\n'.encode())
for container in containers:
    #PRODUCT NAME
    
    product = container.find('div',{'class':'_4rR01T'})
    ProductName = product.text.split('-')[0]
    
   
    #PRODUCT STAR
    
    star = container.find('div',{'class':'_3LWZlK'})
    try:
        Star = star.text
    except:
        Star = 0
        
   #PRODUCT RATING & REVIEWS     
    
    Rating = container.find('span',{'class':'_2_R_DZ'})
    try:
        ratreview= re.findall('\d+,*\d*',Rating.text)
        Ratings = ratreview[0].replace(',','')
        Reviews = ratreview[1].replace(',','')
    except:
        Ratings = 0
        Reviews = 0
    
    
    # PRODUCT CURRENT & MRP PRICE
    
    current_price = container.find('div',{'class':'_30jeq3 _1_WHN1'}).text.replace(',','').replace('','')
    
    mrp = container.find('div',{'class':'_3I9_wc _27UcVY'})
    try:
        MRP = mrp.text.replace(',','').replace('','')
    except:
        MRP = 0
        
    #PRODUCT INFO    
        
    info = container.findAll('li',{'class':'rgWa7D'})
    Processor = info[0].text
    RAM = info[1].text
    Storage = info[3].text
    print(Processor)
    print(RAM)
    print(Storage)
    
    print('\n')
    
  #PRODUCT Image:
    Image = container.img
    ImageURL = Image.get('src')
    
    print(ProductName, Star, Ratings, Reviews, current_price, MRP, Processor, RAM, Storage, ImageURL)
    
    f.write(f"{ProductName},{Star},{Ratings},{Reviews},{current_price},{MRP},{Processor},{RAM},{Storage},{ImageURL}\n".encode())
    print('\n')
    
f.close()   
        
```
